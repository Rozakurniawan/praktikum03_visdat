```python
import matplotlib.pyplot as plt
```

**Dataset**

```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales= [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]
```

**Single Line Plot**  

```python
# Line plot 1 garis, style grafik default
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Multiple Line Plot & Customizations**

```python
from matplotlib.lines import lineStyles
# Kustumisasi line plot
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker="o", color="blue")
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker="x", color="grey")

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukkan Tren**

```python
# Data sample tambahan
product_C_sales = [18, 2, 25, 28, 32, 38, 42, 48, 52, 56, 60, 56]
product_D_sales = [12, 14, 16, 12, 31, 64, 12, 34, 23, 56, 43, 21]

# Plot A
plt.plot(months, product_A_sales, label='Product A', linestyle="--", color="blue")

# Plot B
plt.plot(months, product_B_sales, label='Product B', linestyle="-", color="grey")

# Plot C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color="red")

# Plot D
plt.plot(months, product_D_sales, label='Product D', linestyle="-.", color="green")

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot pada Line Chart**

```python

# Line Plot
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot for Product A
axs[0].plot(months, product_A_sales, label='Product A', color='blue', marker='o')
axs[0].set_title('Penjualan Produk A Per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# Plot for Product B
axs[1].plot(months, product_B_sales, label='Product B', color='red', marker='s')
axs[1].set_title('Penjualan Produk B Per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
